## Cú pháp:

```stata
// Nếu chưa từng được sử dụng bao giờ nhập lệnh dưới đây để tải về: 
ssc install descsave // Không cần chạy nếu đã có trong máy
```

Có nhiều cách ứng dụng câu lệnh descsave trong đó bao gồm việc đổi tên biến, nhãn biến và gán nhãn giá trị cho biến số

Hoặc export thông tin mô tả của bộ dữ liệu để gửi cho đồng nghiệp trong trường hợp cần thiết.

## Ví dụ về đổi tên biến số theo hệ thống từ số liệu xuất ra từ epidata

đây là cách tôi thường xuyên áp dụng cho các bộ dữ liệu nhập bằng epidata

Trên hình là ví dụ 1 bộ dữ liệu sau khi nhập bằng epidata sau đó chuyển sang Stata để phân tích.
Khi xây dựng bộ nhập liệu trên epidata chưa chấp nhận tiếng việt và tên biến số (nhất là đặt tên biến theo phần mục trong bộ câu hỏi) không mang ý nghĩa/thiếu nhãn biến. Do đó, khi chuyển sang Stata cần phải làm cho biến số dễ hiểu, thuận tiện cho việc phân tích.

```stata
use nhaplieu.dta, clear
drop xa hoten dtv // Loại bỏ thông tin định danh của đối tượng điều tra và người phỏng vấn
```

```stata
%head 5
```

Xuất thông tin mô tả bộ dữ liệu bằng lệnh descsave và đổi tên biến, tạo nhãn biến process bằng Excel.

```stata
descsave, saving(thongitn.dta, replace) norestore
```

```stata
%head 5
```

```stata
// Xuất phần thông tin mô tả ra excel để thao tác
export excel using thongtin_mota.xlsx, firstrow(variables) replace
```

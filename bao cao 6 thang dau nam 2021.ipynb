**Lưu ý: trong thực tế có những BN đã đăng ký điều trị và nhập vào máy nhưng không bắt đầu điều trị luôn mà có thể tự chữa sau đó có thể mới quay lại chương trình để điều trị do đó, thời gian tạo trừ ngày bắt đầu điều trị có thể âm. Thậm chí có những trường hợp âm đến vài trăm ngày. Hiện tại không xem kĩ các trường hợp này, có thể là có sai sót trong quá trình nhập liệu nhưng cũng có thể là do thực tế như vậy. Ngoài ra thì có những trường hợp đã điều trị lâu rồi nhưng mà chưa được nhập lên hệ thống, sau đó mới được nhập bổ sung hoặc cũng có thể là do sai sót trong quá trình nhập liệu. Trong khuôn khổ của báo cáo này lấy số trung bình không xem xét từng trường hợp nhỏ lẻ bởi số liệu gửi muộn không đủ thời gian để phân tích kĩ hơn.**

## Mẫu vẽ biểu đồ histogram với đường trung bình đánh dấu

```stata
sysuse nlsw88, clear

summarize wage
local m=r(mean)
local sd=r(sd)
local low = `m'-`sd'
local high=`m'+`sd'

twoway histogram wage , ///
   fc(none) lc(green) xline(`m') ///
   xline(`low', lc(blue)) xline(`high', lc(blue)) scale(0.5) ///
   text(0.12 `m' `"mean = $`=string(`m',"%6.2f")'"', ///
   color(red) orientation(vertical) placement(2))
```

````stata
// hist diff2 if inrange(diff2, -365, 365), freq
qui sum diff2 if inrange(diff2, -365, 365)
local mean_diff2 = r(mean)
twoway histogram diff2 if inrange(diff2, -365, 365), ///
fc(none) lc(blue) xline(`mean_diff2') ///
text(0.12 `mean_diff2' `"Trung bình = `=string(`mean_diff2', "%2.1f")'"', ///
   color(red) orientation(vertical) placement(2))
````

````stata
db twoway
````
